## 🎒 1. O que é uma Lista?
Até agora, nossas variáveis eram como caixinhas que guardavam apenas **um** tesouro por vez. Se você colocasse um item novo, o antigo sumia. 

Uma **Lista** é como um baú do Minecraft ou uma Pokédex: uma única estrutura que pode guardar vários valores ao mesmo tempo, mantendo uma ordem específica.

Imagine um armário. A variável é a etiqueta na porta do armário. Mas, lá dentro, você tem várias prateleiras (compartimentos), e cada uma pode guardar um item diferente.

> [!TIP] Listas vs Vetores:
> Em outras linguagens (como o C), esse conteúdo é chamado de **Vetores**. Normalmente, vetores têm tamanho fixo: se você o cria com 10 espaços, ele terá 10 até o fim. 
> 
> Já no Python, as listas são **dinâmicas**! Elas aumentam e diminuem conforme a sua necessidade. 
> 
> Essa abordagem tem vantagens e desvantagens (como um maior consumo de memória), mas existem bibliotecas no **Python** que implementam operações específicas para vetores, como o **NumPy**.

## 📍 2. Índices: O Endereço do Item
Cada item em uma lista tem uma posição exata, que chamamos de **Índice**. Voltando à ideia do armário, é como se cada gaveta tivesse um número. Assim, conseguimos pegar exatamente o que queremos:


```python
# Criando uma lista de iniciais
iniciais: list[str] = ["Bulbasaur", "Charmander", "Squirtle"]

# Acessando pelo índice
print(iniciais[0]) # Bulbasaur
print(iniciais[2]) # Squirtle
```


> [!WARNING] REGRA ZERO DO PROGRAMADOR:
> O computador sempre começa a contar do **0**, não do 1! Se sua lista tem 3 itens, os endereços são 0, 1 e 2.

### ⬅️ Índices Negativos
Outro recurso que não é comum em outras linguagens é a função de índices negativos. O Python aceita contar de trás para frente:
- `lista[-1]` é sempre o último.
- `lista[-2]` é o penúltimo.


```python
print(iniciais[-1]) # Squirtle
```

## 🛠️ 3. Manipulando a Mochila (Métodos)
Para gerenciar nossa lista, usamos "funções especiais" que vêm grudadas nelas:

| Comando              | O que faz?                             | Exemplo                      |
| :------------------- | :------------------------------------- | :--------------------------- |
| `.append(item)`      | Adiciona ao **final** da lista.        | `equipe.append("Pikachu")`   |
| `.insert(pos, item)` | Coloca em uma posição **específica**.  | `equipe.insert(0, "Mewtwo")` |
| `.pop(pos)`          | Remove e "entrega" o item da posição.  | `item = mochila.pop(2)`      |
| `.remove(item)`      | Procura e remove o item pelo **nome**. | `mochila.remove("Pedra")`    |
| `len(lista)`         | Conta quantos itens existem.           | `total = len(equipe)`        |


```python
inventario: list[str] = ["Espada", "Picareta"]
inventario.append("Tocha") # ["Espada", "Picareta", "Tocha"]
inventario[0] = "Espada de Diamante" # Substituindo o item 0
```

## 🔄 4. Listas + Loops: O Combo Perfeito
A forma mais poderosa de usar listas é percorrendo-as com o loop `for`. Temos dois jeitos principais:

### A) Pelo Índice (Usando o endereço)
Usamos a função `len()` para saber o tamanho e o `range()` para criar os números dos endereços:


```python
equipe: list[str] = ["Charmander", "Pikachu", "Mew"]
tamanho: int = len(equipe)

for i in range(tamanho):
    print(f"Slot {i}: {equipe[i]}")
```


Note que o `range(5)` cria uma sequência `[0, 1, 2, 3, 4]`. Por isso ele combina perfeitamente com os índices da lista! Nós construímos uma lista de todos os compartimentos do armário e usamos ela para acessar.

A função `range()` na verdade também pode ser convertida (cast) em uma lista:


```python
numeros: list[int] = list(range(5))
print(numeros)
```

### B) Pelo Item
Na aula anterior vimos que o loop `for` continua **enquanto existirem elementos** na fila:


```python
equipe: list[str] = ["Pikachu", "Eevee", "Ditto"]
for pokemon in equipe:
    print(f"Eu escolho você, {pokemon}!")
```

## 🔍 5. Acessórios de Exploração
Às vezes, só listar os itens não é o bastante. Precisamos de ferramentas mais avançadas para gerenciar nosso inventário!
### `Enumerate`:
O `enumerate` é como se você estivesse olhando para sua prateleira e visse, ao mesmo tempo, o número da etiqueta e o objeto que está lá. Ele nos entrega o **índice** e o **valor** de uma vez só!


```python
itens: list[str] = ["Mapa", "Bússola", "Poção"]

# i recebe o índice, item recebe o nome do objeto
for i, item in enumerate(itens):
	print(f"Slot {i}: {item}")
```


### `Zip`: Emparelhando Listas
Às vezes, temos duas listas que andam juntas, como se fossem as duas metades de algo. 
- **Por exemplo:** uma lista com nomes de jogadores e outra com suas respectivas pontuações.

O `zip()` serve para juntar essas duas listas e percorrê-las ao mesmo tempo:


```python
jogadores: list[str] = ["Steve", "Alex", "Creeper"]
pontos: list[int] = [100, 250, 50]

for nome, pontuacao in zip(jogadores, pontos):
    print(f"O jogador {nome} tem {pontuacao} pontos.")
```


### O Operador `in`
Quer saber se um item está na mochila sem precisar olhar gaveta por gaveta? Use o `in`:


```python
inventario: list[str] = ["Terra", "Diamante"]

if "Diamante" in inventario:
	print("Você está rico! 💎")
```
